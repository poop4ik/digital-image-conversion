### Імпорт модулів

### 1. Робота з чорно-білим фото

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('IMG10.BMP', cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Помилка: зображення не знайдено!")
else:
    # Побудова гістограми оригінального зображення
    hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
    
    # Еквалізація гістограми
    equalized_image = cv2.equalizeHist(image)
    hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
    
    # Відображення результатів
    plt.figure(figsize=(12, 6))
    
    # Оригінальне зображення
    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Оригінальне зображення')
    plt.axis('off')
    
    # Гістограма оригінального зображення
    plt.subplot(2, 2, 2)
    plt.plot(hist_original, color='black')
    plt.title('Гістограма оригінального зображення')
    
    # Еквалізоване зображення
    plt.subplot(2, 2, 3)
    plt.imshow(equalized_image, cmap='gray')
    plt.title('Еквалізоване зображення')
    plt.axis('off')
    
    # Гістограма еквалізованого зображення
    plt.subplot(2, 2, 4)
    plt.plot(hist_equalized, color='black')
    plt.title('Гістограма еквалізованого зображення')
    
    plt.tight_layout()
    plt.show()
```

### 2. Робота з кольоровим фото

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('IMG10.BMP')

if image is None:
    print("Помилка: зображення не знайдено!")
else:
    # Перетворення зображення в формат RGB для відображення
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Розділення зображення на канали
    b, g, r = cv2.split(image)

    # Побудова гістограм для кожного каналу
    hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
    hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

    # Еквалізація кожного каналу
    b_eq = cv2.equalizeHist(b)
    g_eq = cv2.equalizeHist(g)
    r_eq = cv2.equalizeHist(r)

    # З'єднання еквалізованих каналів
    equalized_image = cv2.merge([b_eq, g_eq, r_eq])
    equalized_image_rgb = cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB)

    # Побудова гістограм для еквалізованих каналів
    hist_b_eq = cv2.calcHist([b_eq], [0], None, [256], [0, 256])
    hist_g_eq = cv2.calcHist([g_eq], [0], None, [256], [0, 256])
    hist_r_eq = cv2.calcHist([r_eq], [0], None, [256], [0, 256])

    # Відображення результатів
    plt.figure(figsize=(12, 6))

    # Оригінальне зображення
    plt.subplot(2, 2, 1)
    plt.imshow(image_rgb)
    plt.title('Оригінальне зображення')
    plt.axis('off')

    # Гістограма оригінального зображення
    plt.subplot(2, 2, 2)
    plt.plot(hist_b, color='blue', label='Синій')
    plt.plot(hist_g, color='green', label='Зелений')
    plt.plot(hist_r, color='red', label='Червоний')
    plt.title('Гістограма оригінального зображення')
    plt.legend()

    # Еквалізоване зображення
    plt.subplot(2, 2, 3)
    plt.imshow(equalized_image_rgb)
    plt.title('Еквалізоване зображення')
    plt.axis('off')

    # Гістограма еквалізованого зображення
    plt.subplot(2, 2, 4)
    plt.plot(hist_b_eq, color='blue', label='Синій')
    plt.plot(hist_g_eq, color='green', label='Зелений')
    plt.plot(hist_r_eq, color='red', label='Червоний')
    plt.title('Гістограма еквалізованого зображення')
    plt.legend()

    plt.tight_layout()
    plt.show()
```
