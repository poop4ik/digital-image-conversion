### Імпорт модулів

### 1. Робота з чорно-білим фото

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Завантаження зображень
image1 = cv2.imread('IMG10.BMP', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('myphoto.jpg', cv2.IMREAD_GRAYSCALE)

if image1 is None or image2 is None:
    print("Помилка: одне з зображень не знайдено!")
else:
    # Обробка першого зображення
    hist_original1 = cv2.calcHist([image1], [0], None, [256], [0, 256])
    equalized_image1 = cv2.equalizeHist(image1)
    hist_equalized1 = cv2.calcHist([equalized_image1], [0], None, [256], [0, 256])
    
    # Обробка другого зображення
    hist_original2 = cv2.calcHist([image2], [0], None, [256], [0, 256])
    equalized_image2 = cv2.equalizeHist(image2)
    hist_equalized2 = cv2.calcHist([equalized_image2], [0], None, [256], [0, 256])
    
    # Відображення результатів
    plt.figure(figsize=(8, 8))  # Зменшення розміру фігури
    
    # Оригінальне зображення 1
    plt.subplot(2, 2, 1)
    plt.imshow(image1, cmap='gray')
    plt.title('Оригінальне зображення 1')
    plt.axis('off')
    
    # Гістограма оригінального зображення 1
    plt.subplot(2, 2, 2)
    plt.plot(hist_original1, color='black')
    plt.title('Гістограма оригінального зображення 1')
    
    # Еквалізоване зображення 1
    plt.subplot(2, 2, 3)
    plt.imshow(equalized_image1, cmap='gray')
    plt.title('Еквалізоване зображення 1')
    plt.axis('off')
    
    # Гістограма еквалізованого зображення 1
    plt.subplot(2, 2, 4)
    plt.plot(hist_equalized1, color='black')
    plt.title('Гістограма еквалізованого зображення 1')
    
    plt.tight_layout()
    plt.show()
    
    # Для другого зображення відображаємо в окремому вікні
    plt.figure(figsize=(8, 8))  # Зменшення розміру фігури
    
    # Оригінальне зображення 2
    plt.subplot(2, 2, 1)
    plt.imshow(image2, cmap='gray')
    plt.title('Оригінальне зображення 2')
    plt.axis('off')
    
    # Гістограма оригінального зображення 2
    plt.subplot(2, 2, 2)
    plt.plot(hist_original2, color='black')
    plt.title('Гістограма оригінального зображення 2')
    
    # Еквалізоване зображення 2
    plt.subplot(2, 2, 3)
    plt.imshow(equalized_image2, cmap='gray')
    plt.title('Еквалізоване зображення 2')
    plt.axis('off')
    
    # Гістограма еквалізованого зображення 2
    plt.subplot(2, 2, 4)
    plt.plot(hist_equalized2, color='black')
    plt.title('Гістограма еквалізованого зображення 2')
    
    plt.tight_layout()
    plt.show()
```
